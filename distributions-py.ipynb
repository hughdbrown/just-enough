```python
# Scipy.stats has loads of distributions that you can use. Importing scipy.stats as scs is the 
# norm, and you should follow it. All of these others below are the norm, too. 
import scipy.stats as scs
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# We can initialize a normal distrbution as follows. 
norm = scs.norm()
```

```python
# We can get 1000 random variates as follows. 
normed_values = norm.rvs(size = 1000)
```

```python
# We can get the pdf for a given set of x's as follows. 
Xs = np.linspace(start = -3, stop = 3, num = 1000)
Ys = norm.pdf(Xs)
```

```python
# Let's plot them and see what they look like. 
plt.hist(normed_values)
plt.show()
plt.plot(Xs, Ys)
plt.show()
```

```python
# With these distributions, we can specify what we want the parameters to be. For the norm, 
# these parameters are the mean and standard deviation. 
norm2 = scs.norm(loc = 5, scale = 5)
normed2_values = norm2.rvs(size = 1000)
Xs2 = np.linspace(start = -15, stop = 20, num = 1000)
Ys2 = norm2.pdf(Xs2)
```

```python
# Let's plot this distribution to see what it looks like. It should be centered around 5, with much wider 
# tails than those above. 
plt.hist(normed2_values)
plt.show()
plt.plot(Xs2, Ys2)
plt.show()
```

```python
# We can also get random variates for a binomial distribution as well. This procedure is exactly the same 
# as for the normal distribution; we just have to make sure to put in the right parameters. For the binomial, 
# these parameters are n and p. 
binom = scs.binom(n = 100, p = 0.5)
binom_values = binom.rvs(size = 1000)
plt.hist(binom_values)
plt.show()
```
